```python
import pandas as pd

from bokeh.models import ColumnDataSource
from bokeh.plotting import figure, show, output_notebook
from bokeh.tile_providers import get_provider, Vendors
from pathlib import Path
from pyproj import Proj, transform

"""Initialize geo map"""
proj = Proj(proj="webmerc")

longitude_range = (-75.239259, -75.004820)
latitude_range = (39.863319, 40.094042)

merc_x, merc_y = proj(longitude_range, latitude_range)

p = figure(x_range=merc_x, y_range=merc_y,
           x_axis_type="mercator", y_axis_type="mercator")
p.add_tile(get_provider(Vendors.CARTODBPOSITRON_RETINA))

"""Load CSV data"""
psp_overdose_events = pd.read_csv(os.getcwd() + "/data/psp_overdose_events.csv")

"""Add relevant CSV data to geo map"""
incident_merc_x, incident_merc_y = proj(psp_overdose_events["Incident County Longitude"].tolist(), 
                                        psp_overdose_events["Incident County Latitude"].tolist())

p.circle(x=incident_merc_x, y=incident_merc_y)

output_notebook()
show(p)
```
