```python
kept_od_inc_zips = dph_od_incident_counts[~dph_od_incident_counts['ZIPCODE'].isin(['00J', 'Unknown'])]
kept_od_res_zips = dph_od_resident_counts[~dph_od_resident_counts['Zip.Code'].isin(['00J', 'Unknown'])]
```

- for simplicity, initial version is to first assign all nonzero incident/resident overdose counts as 1, whilst zero as 0.
- this boolean flagging approach will give us a rough sense of directionally how overdose locations move over time.

- current version- hardcoded threshold -> 10 to create mask.
- this effectively marks the regions which have 10 or more incidents per zip code while ignoring the rest.
- myopic in view, but intention is to see if high prevlance zips (as previously defined by threshold) change over time.

- Afterward, we can refine and pinpoint on top of this to show where the <6 incidents or higher are occurring

```python
kept_od_inc_zips = kept_od_inc_zips.replace('<6', '0')
kept_zips = kept_od_inc_zips['ZIPCODE']
kept_zip_vals = kept_od_inc_zips.loc[:, kept_od_inc_zips.columns != 'ZIPCODE']

kept_zip_vals_numeric = kept_zip_vals.apply(pd.to_numeric)

kept_od_inc_zips = pd.concat([kept_zips, kept_zip_vals_numeric], axis=1)
```

```python
# boolean flagging by overdose occurrence or not

def bool_flag_overdose_incidents(df, zipcode_col, threshold=0):
    """
    convert overdose counts above threshold to 1, whilst below threshold counts to 0
    
    * README:
    - not an automated function to use in bulk, just an exploration
    
    * current logic relies on creating a boolean mask to flag zip codes as areas which have
    incident counts above a certain threshold
    """
    vals = df[df.columns[df.columns != zipcode_col]]
    
    thresh_mask = (vals > threshold)
    bool_vals = vals.where(thresh_mask, other=0)

    df = pd.concat([df[zipcode_col], bool_vals], axis=1)
    
    return df 

bool_inc_zip_cts = bool_flag_overdose_incidents(kept_od_inc_zips, 'ZIPCODE', threshold=0)
# bool_res_zip_cts = bool_flag_overdose_incidents(kept_od_res_zips, 'Zip.Code')
```

```python
# execute for loop to create new folium Map() object and populate with the choropleth map data per iteration for every quarter

bool_inc_zip_cts['ZIPCODE']=bool_inc_zip_cts['ZIPCODE'].astype(str)

quarters = bool_inc_zip_cts.columns[1:]

for q in quarters:
    print('quarter:', q)
    map = fm.Map(location=[39.9526, -75.1652], default_zoom_start=15)
    
    map.choropleth(titles="Quarter {0}".format(q), geo_data=basepath+"Zipcodes_Philly_Polygon.geojson",
                 data=bool_inc_zip_cts[['ZIPCODE', q]], # my dataset
                 columns=['ZIPCODE', q],
                 key_on='feature.properties.CODE',
                 fill_color='BuPu', fill_opacity=0.4, line_opacity=0.4,
                 legend_name='Incident Overdose Counts by Zip, (incidents > 6)')

    # html file name
    html_filename = basepath+'output/{0}_Incident_Overdose_Occurrences_sup6.html'.format(q)
    
    map.save(html_filename)
    
    # use selenium to save .html to .png
#     print('html filename:', html_filename)
    
    png_filename = basepath+'output/{0}_Incident_Overdose_Occurrences_sup6.png'.format(q)
#     print('png filename:', png_filename)
    
    # use selenium to scrape from loaded HTML page
    driver = selenium.webdriver.PhantomJS(executable_path='/usr/local/bin/phantomjs')
    driver.set_window_size(600, 400)  # choose a resolution
    driver.get(html_filename)
    
    # You may need to add time.sleep(seconds) here
    time.sleep(3)
    
    driver.save_screenshot(png_filename)
    
```
